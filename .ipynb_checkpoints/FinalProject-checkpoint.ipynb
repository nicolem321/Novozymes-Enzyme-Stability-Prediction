Compare sequences to see if they differ by less than 2 mutation

```python
def compare_seq(seq1,seq2):
    if(type(seq1)!=str):
        return 0
    if(type(seq2)!=str):
        return 0
    if(len(seq1)>len(seq2)+2 or len(seq1)<len(seq2)-2):
        return 0
    #print('seq1 len',len(seq1))
    count=0
    minlen=min(len(seq1),len(seq2))
    for i in range(minlen):
        if seq1[i]!=seq2[i]:
            count=count+1
    if count>3:
        return 0
    #print(count)
    return 1
```

```python
def is_subseq(seq1, seq2):
    it = iter(seq2)
    return all(any(c1 == c2 for c1 in it) for c2 in seq1)
```

## set wild type sequence

```python
def avg_len(seqs):
    sum=0
    for seq in seqs:
        sum+=len(seq)
    return sum//len(seqs)
```

```python
def most_seen(ac_list):
    #input: a list of amino acids that appeared in the same position acorss different proteins of the same type
    #output: the most seen amino acids at this position, return it so we can add it to its wildtype at this position
    dictionary = dict()
    for i in range(0, len(ac_list)):
        ac = ac_list[i]
        if dictionary.__contains__(ac):
            #ac already in, increase count
            count = dictionary[ac]
            count += 1
            dictionary[ac] = count
        else:
            #we have not seen this, initialize count
            count = 1
            dictionary[ac] = count
            
    #now we have a dictionary of keys and their counts
    max = 0
    most_ac = None
    for key in dictionary.keys():
        if dictionary[key] > max:
            max = dictionary[key]
            most_ac = key
    return most_ac
```

```python
#first, we will have a dictionary
dictionary = dict() #keys are unique group ids, values are lists of protein sequences in this group

for i in range(0, 6794):
    #iterating through all the rows in dataframe
    group_id = train.iloc[i]["group"]
    protein_sequence = train.iloc[i]["protein_sequence"]
    if dictionary.__contains__(group_id):
        new_list = dictionary[group_id]
        new_list.append(protein_sequence)
        #we will append it to the end of the list
        dictionary[group_id] = new_list
    else:
        #we haven't seen this group id before
        #we want to put it in the dictionary
        protein_list = []
        protein_list.append(protein_sequence)
        dictionary[group_id] = protein_list
```

```python
for group_id in dictionary.keys():
#give me all the group ids 
    wildtype = "" #represents the wildtype of this group
    seqs = dictionary[group_id]
    wildtype_length = avg_len(seqs) #taking the average length 
    ac_dict = {}
    for seq in seqs:
        #strings of particular seq
        for i in range(0, len(seq)):
            if i > wildtype_length:
                break;
            #accessing indivual char in string seq
            char = seq[i]
            if ac_dict.__contains__(i):
                new_list = ac_dict[i]
                new_list.append(char)
                #we will append it to the end of the list
                ac_dict[i] = new_list
            else:
                #we have not seen the position
                ac_list = []
                ac_list.append(char)
                ac_dict[i] = ac_list
                
    #by this point, we will have ac_dict, which will contain position as key, ac_list as value
    for key in ac_dict.keys():
        position_list = ac_dict[key]
        ac = most_seen(position_list)
        wildtype = wildtype + ac
        
    dictionary[group_id] = wildtype
```

# Model 1: Gradient Boosting Regression Tree Model

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error, f1_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
import scipy
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

#transform the columns that are objects(protein sequence and wildtype)
cat_cols = [col_name for col_name in X_train_x.columns if X_train_x[col_name].dtype == "object"]
cat_trans = Pipeline(steps=[
           ('imputer', SimpleImputer(strategy='constant', fill_value='unknown')),
           ('onehot', OneHotEncoder(handle_unknown='ignore'))])
preprocessor = ColumnTransformer(transformers=[('cat', cat_trans, cat_cols)])

#define our model
regressor=GradientBoostingRegressor(max_depth=2,
                                    n_estimators=3,
                                    learning_rate=1)

model_pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('model', regressor)])

model_pipeline.fit(X_train_x,X_train_y)

y_pred=model_pipeline.predict(X_test_x) #the prediction

print(y_pred)
for i in y_pred:
    print (i)

print("The Pearson correlation coefficient, PCC, for GBRT is", scipy.stats.pearsonr(X_test_y, y_pred)[0])
print("The root mean square deviation, RMSD, for GBRT is", mean_squared_error(X_test_y, y_pred, squared=False))
print("The R2 for GBRT is", r2_score(X_test_y, y_pred))
print("The mean squared error, MSE, for GBRT is", mean_squared_error(X_test_y, y_pred))
print("The mean absolute error, MAE, for GBRT is", mean_absolute_error(X_test_y, y_pred))
```

# Model 2: XGBoost

```python
from scipy.stats import spearmanr
import scipy
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.pipeline import Pipeline
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from numpy import arange
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
```

```python
#transform the columns that are objects(protein sequence and wildtype)
cat_cols = [col_name for col_name in X_train_x.columns if X_train_x[col_name].dtype == "object"]
cat_trans = Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='constant', fill_value='unknown')),
            ('onehot', OneHotEncoder(handle_unknown='ignore'))])
preprocessor = ColumnTransformer(transformers=[('cat', cat_trans, cat_cols)])

model = XGBRegressor(scale_pos_weight=1,
                     learning_rate=0.01,
                     colsample_bytree=0.4,
                     subsample=0.8,
                     objective='reg:squarederror',
                     eval_metric=mean_absolute_error,
                     n_estimators=1000,
                     reg_alpha=0.3,
                     max_depth=4,
                     gamma=10)

model_pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('model', model)])

model_pipeline.fit(X_train_x, X_train_y)

xg_y_pred = model_pipeline.predict(X_test_x)

print(xg_y_pred)
for i in xg_y_pred:
    print (i)

print("The Pearson correlation coefficient, PCC, for XGBoost is", scipy.stats.pearsonr(X_test_y, xg_y_pred)[0])
print("The root mean square deviation, RMSD, for XGBoost is", mean_squared_error(X_test_y, xg_y_pred, squared=False))
print("The R2 for XGBoost is", r2_score(X_test_y, xg_y_pred))
print("The mean squared error, MSE, for XGBoost is", mean_squared_error(X_test_y, xg_y_pred))
print("The mean absolute error, MAE, for XGBoost is", mean_absolute_error(X_test_y, xg_y_pred))
```
